step 1: Use yfinance.download() directly

```python
import yfinance as yf
import pandas as pd

tickers = ["SPY", "AGG", "VNQ", "DBC"]
start   = "2018-01-01"
end     = "2024-12-31"

data = yf.download(tickers, start=start, end=end, progress=False)

# If it's a MultiIndex (multiple tickers), first level has names like 'Adj Close'
if isinstance(data.columns, pd.MultiIndex):
    if "Adj Close" in data.columns.get_level_values(0):
        prices = data["Adj Close"]
    else:
        prices = data["Close"]
# If it's a flat Index (single ticker), try both
else:
    if "Adj Close" in data.columns:
        prices = data["Adj Close"]
    else:
        prices = data["Close"]

returns = prices.pct_change().dropna()

print(prices.tail())
print(returns.tail())


```

Step 2 - Compute Basic Statistics

```python
import numpy as np
import pandas as pd

# After you’ve loaded `prices` and computed `returns`, we want to be able to compute amnt of trading days from start to finish below and give us our simple stats

# 1) Compute actual trading days between start & end:
trading_days = len(pd.bdate_range(start=start, end=end))

# 2) Daily stats
mean_daily_ret = returns.mean()
cov_daily      = returns.cov()

# 3) Annualize using the actual trading_days
mean_annual_ret = mean_daily_ret * trading_days
cov_annual      = cov_daily * trading_days

print(f"Trading days in sample: {trading_days}")
print("Annualized mean returns:\n", mean_annual_ret)
print("\nAnnualized covariance matrix:\n", cov_annual)
```

Step 3 - Monte Carlo Simulation & Efficent Frontier using the amount of trading days

```python
import numpy as np
import matplotlib.pyplot as plt

# Assume mean_annual_ret and cov_annual are already defined from before and trading_days = 1827 (from previous output)

n_portfolios = 5_000
n_assets     = len(mean_annual_ret)
results = np.zeros((3, n_portfolios))

for i in range(n_portfolios):
    # 1) Random weights
    w = np.random.random(n_assets)
    w /= np.sum(w)
    
    # 2) Portfolio return & volatility
    port_ret = w @ mean_annual_ret
    port_vol = np.sqrt(w.T @ cov_annual @ w)
    
    # 3) Record vol, ret, Sharpe (rf=0)
    results[0, i] = port_vol
    results[1, i] = port_ret
    results[2, i] = port_ret / port_vol

# Plot Monte Carlo cloud
plt.figure()
plt.scatter(results[0], results[1], s=8, alpha=0.3)
plt.xlabel('Annualized Volatility')
plt.ylabel('Annualized Return')
plt.title(f'Monte Carlo Efficient Frontier ({trading_days} trading days)')
```

# Choosing the Right Optimizer for Mean–Variance Portfolios

For a portfolio with box‐bounds on each weight (\(0 \le w_i \le 1\)) and the equality constraint \(\sum_i w_i = 1\), these solvers are the most relevant:

| Solver         | Bounds? | Equality? | Speed     | Robustness                            |
|----------------|:-------:|:---------:|:---------:|:--------------------------------------|
| **SLSQP**      | ✔️      | ✔️        | High      | Default choice; easy & reliable       |
| **trust-constr** | ✔️    | ✔️        | Medium    | Very robust; handles tricky cases     |
| **L-BFGS-B**   | ✔️      | ❌        | Very High | Fast but needs reparameterization     |
| **COBYLA**     | ✔️ via constraints | ✔️ | Low       | Derivative-free; slower convergence   |

> **Quick Summary:**  
> - **Start with SLSQP** for small, smooth problems with both bounds and \(\sum w = 1\).  
> - If SLSQP stalls or you need higher reliability, switch to **trust-constr**.  
> - Use **L-BFGS-B** only if you reparameterize (e.g., softmax) or add a penalty for \(\sum w=1\).  
> - Try **COBYLA** or **Powell** as a last resort when gradients aren’t available.


```python
from scipy.optimize import minimize

def portfolio_vol(w, cov):
    return np.sqrt(w.T @ cov @ w)

def neg_sharpe(w, mean_ret, cov):
    ret = w @ mean_ret
    vol = portfolio_vol(w, cov)
    return -ret / vol

constraints = ({'type': 'eq', 'fun': lambda w: np.sum(w) - 1})
bounds      = tuple((0, 1) for _ in range(n_assets))
init_w      = np.repeat(1/n_assets, n_assets)

opt = minimize(
    neg_sharpe, 
    init_w, 
    args=(mean_annual_ret, cov_annual),
    method='SLSQP', 
    bounds=bounds, 
    constraints=constraints
)

opt_w = opt.x
opt_ret = opt_w @ mean_annual_ret
opt_vol = portfolio_vol(opt_w, cov_annual)

# Overlay optimized point
plt.scatter(opt_vol, opt_ret, marker='*', s=200, color='red')
plt.show()

print("Max-Sharpe weights:")
for t, weight in zip(mean_annual_ret.index, opt_w):
    print(f"  {t}: {weight:.2%}")
```
